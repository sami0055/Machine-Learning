```python
df=df.iloc[:,2:]
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OrdinalEncoder
```

```python
X_train,X_test,y_train,y_test=train_test_split(df.iloc[:,0:2],df.iloc[:,-1],test_size=0.2,random_state=42)
```

```python
X_train
```

```python
X_test.head(10)
```

```python
od=OrdinalEncoder(categories=[['Poor','Average','Good'],['School','UG','PG']])
```

```python
X_train=od.fit_transform(X_train)
X_test=od.transform(X_test)
```

```python
print(X_test)
```

```python
od.categories
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
```

```python
y_train=le.fit_transform(y_train)
```

```python
y_test=le.transform(y_test)
```
