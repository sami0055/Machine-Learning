# Which regions have more populations?

```python
#Numerical Analysis
grouped=df.groupby("Region")
agg=grouped.agg({'Population_mln':['mean']})
agg
```

```python

sns.barplot(data=df,y='Region',x='Population_mln')
plt.xlabel("Population in Million")
plt.ylabel("Regions")
plt.title("Regions V/S Populations in Million")
plt.yticks(rotation=45)
plt.show()
```

### Here it is clear that the highest population in North America, the second highest in Asia, and the region with the lowest population is Oceania.

# Which regions are more affected by HIV incidents?

```python
grouped=df.groupby('Region')
agg=grouped.agg({'Incidents_HIV':['mean']})
agg
```

```python
sns.barplot(data=df,y='Region',x='Incidents_HIV')
plt.xlabel("Incidents of HIV")
plt.ylabel("Region")
plt.title("Region V/S HIV Incidents")
plt.yticks(rotation=45)
plt.show()
```

```python

grouped=df.groupby('Region')
agg=grouped.agg({'Incidents_HIV':['mean']})
arr=np.array([])
regions=np.array([])
for i,value in agg.iterrows():
    regions=np.append(regions,i)
    arr=np.append(arr,value)


fig, ax = plt.subplots(figsize=(5, 5))  # Adjust figure size as needed
wedges, texts, autotexts = ax.pie(arr, autopct='%1.1f%%', startangle=90, pctdistance=0.8)

# Adjust label and autopct distances to avoid overlap
for text, autotext in zip(texts, autotexts):
    text.set_size('smaller')
    autotext.set_size('smaller')

# Make the legend outside the pie chart
ax.legend(wedges, regions, title="Regions",  loc="center left", bbox_to_anchor=(1, 0.5),bbox_transform=plt.gcf().transFigure)

# Equal aspect ratio ensures that pie is drawn as a circle
ax.axis('equal')

plt.show()
```

### From this two graph, we saw that the Aferica region is the most affected by HIV incidents, and the second highest is Central America and Caribbean. The lowest one is the Europian Union.
